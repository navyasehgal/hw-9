<div class="alert alert-warning">
    
Solution_4

````python
# Feature engineering

# Average bill amount across 6 months
X_train["AVG_BILL_AMT"] = X_train[["BILL_AMT1", "BILL_AMT2", "BILL_AMT3",
                                   "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]].mean(axis=1)

# Average payment amount across 6 months
X_train["AVG_PAY_AMT"] = X_train[["PAY_AMT1", "PAY_AMT2", "PAY_AMT3",
                                  "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"]].mean(axis=1)

# Repayment status trend
pay_cols = ["PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6"]
X_train["PAY_TREND"] = X_train[pay_cols].mean(axis=1)

# Utilization ratio: last month's bill / credit limit
X_train["UTILIZATION_RATIO"] = X_train["BILL_AMT1"] / (X_train["LIMIT_BAL"] + 1e-6)
````

<!-- END QUESTION -->

<br><br>

For this dataset, I created a small set of meaningful engineered features that capture financial behavior trends that are not directly represented in the raw columns.

New features created

AVG_BILL_AMT: The average bill amount over the last six months.
This captures the overall spending pattern of each client more clearly than looking at individual bill statements.

AVG_PAY_AMT: The average payment amount over the last six months.
This summarizes how consistently a client makes payments relative to their bills.

PAY_TREND: The overall trend in repayment status (mean of PAY_0 to PAY_6).
Higher values indicate worsening delays, which should be predictive of default.

UTILIZATION_RATIO: The ratio of most recent bill amount to credit limit.
This approximates how much of their credit a client is using, which is an important risk factor.

These engineered features capture repayment behavior, credit usage, and spending patterns more effectively than individual raw columns, and will be included in the modeling pipeline to help improve predictive performance.

````python
# Identify feature types

numeric_features = [
    "LIMIT_BAL", "AGE",
    "BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6",
    "PAY_AMT1", "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6",
    "AVG_BILL_AMT", "AVG_PAY_AMT", "PAY_TREND", "UTILIZATION_RATIO"]

categorical_features = [
    "SEX", "EDUCATION", "MARRIAGE",
    "PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6"]
````

````python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define preprocessing pipeline

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)])
````

I identified numerical features such as bill amounts, payment amounts, engineered features, and credit limit, which benefit from scaling. 
Categorical variables including demographic fields (SEX, EDUCATION, MARRIAGE) and payment status codes (PAY_0 to PAY_6) were assigned to one-hot encoding. 
I used a ColumnTransformer to apply StandardScaler to numerical features and OneHotEncoder to categorical features, ensuring consistent preprocessing before modeling.

I trained a logistic regression model using a pipeline that included all of my preprocessing steps and class weighting to address the imbalance in the default variable. I performed hyperparameter tuning on the regularization strength C using a grid search over four values: 0.01, 0.1, 1, and 10. The model was evaluated using five fold cross validation with the F1 weighted metric.

The results showed consistent performance across folds, with the following mean F1 weighted scores for each C value:

````python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

# Create a pipeline: preprocessing + logistic regression
log_reg_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", LogisticRegression(max_iter=2000, class_weight="balanced"))])

# Hyperparameter grid for regularization strength
param_grid = {
    "model__C": [0.01, 0.1, 1, 10]}

# Grid search with cross validation
grid_lr = GridSearchCV(
    log_reg_pipeline,
    param_grid,
    cv=5,
    scoring="f1_weighted",
    n_jobs=-1)

grid_lr.fit(X_train, y_train)

lr_results = grid_lr.cv_results_
best_lr_score = grid_lr.best_score_
best_lr_params = grid_lr.best_params_

best_lr_score, best_lr_params
````

````python
# RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

rf_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(random_state=123, n_jobs=-1))])

param_grid_rf = {
    "model__n_estimators": [100, 200],
    "model__max_depth": [None, 10, 20],
    "model__min_samples_split": [2, 5]}

grid_rf = GridSearchCV(
    rf_pipeline,
    param_grid_rf,
    cv=5,
    scoring="f1_weighted",
    n_jobs=-1)

grid_rf.fit(X_train, y_train)

print("Random Forest Results:")
print("Best CV Score:", grid_rf.best_score_)
print("Std Dev:", grid_rf.cv_results_["std_test_score"][grid_rf.best_index_])
print("Best Params:", grid_rf.best_params_)
````

````python
# GRADIENT BOOSTING
from sklearn.ensemble import GradientBoostingClassifier

gb_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", GradientBoostingClassifier(random_state=123))])

param_grid_gb = {
    "model__n_estimators": [100, 200],
    "model__learning_rate": [0.05, 0.1],
    "model__max_depth": [2, 3]}

grid_gb = GridSearchCV(
    gb_pipeline,
    param_grid_gb,
    cv=5,
    scoring="f1_weighted",
    n_jobs=-1)

grid_gb.fit(X_train, y_train)

print("Gradient Boosting Results:")
print("Best CV Score:", grid_gb.best_score_)
print("Std Dev:", grid_gb.cv_results_["std_test_score"][grid_gb.best_index_])
print("Best Params:", grid_gb.best_params_)
````

````python
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

# LinearSVC pipeline
lsvc_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", LinearSVC(class_weight="balanced", max_iter=5000))])

# Hyperparameter grid
param_grid_lsvc = {
    "model__C": [0.01, 0.1, 1, 10]}

# Grid search
grid_lsvc = GridSearchCV(
    lsvc_pipeline,
    param_grid_lsvc,
    cv=5,
    scoring="f1_weighted",
    n_jobs=-1)

grid_lsvc.fit(X_train, y_train)

print("LinearSVC Results:")
print("Best CV Score:", grid_lsvc.best_score_)
print("Best Params:", grid_lsvc.best_params_)
````
